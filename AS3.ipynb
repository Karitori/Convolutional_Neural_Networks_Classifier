```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import preprocessing
from numpy.random import seed
from skimage.io import imread_collection
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from skimage.io import imread_collection
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.environ['PYTHONHASHSEED'] = '0'
np.random.seed(37)
tf.random.set_seed(89)
train_new_model = True
```

```python
def accuracy(true,result):
    score = (accuracy_score(true,result)*100)
    print("Accuracy: ",score,"%")
    # return score
    
def confusion(true,result):
    cm = confusion_matrix(true,result)
    print(cm)
    # return cm
```

```python
images = np.array(imread_collection(r'./Train/*.jpg'))

train=np.zeros((2400, 784))
for i in range(len(images)):
    train[i]=images[i].flatten()
```

```python
test_images = np.array(imread_collection(r'./Test/*.jpg'))
test=np.zeros((200, 784))
for i in range(len(test_images)):
    test[i]=test_images[i].flatten()
```

```python
labels = np.loadtxt(r"./Train/Training Labels.txt", dtype=int)
labels_test = np.loadtxt(r"./Test/Test Labels.txt", dtype=int)
```

```python
X_train = tf.convert_to_tensor(tf.constant(train, dtype=tf.float32))
Y_train = tf.keras.utils.to_categorical(labels, 10)
standard_train = preprocessing.StandardScaler().fit_transform(X_train)
```

```python
X_test = tf.convert_to_tensor(tf.constant(test , dtype = tf.float32))
Y_test = tf.keras.utils.to_categorical(labels_test, 10)
standard_test = preprocessing.StandardScaler().fit_transform(X_test)
```

```python
#First Requirement

model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dense(units=500, activation=tf.nn.relu))
model1.add(tf.keras.layers.Dense(units=10, activation=tf.nn.sigmoid))
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mean_squared_error',metrics=['binary_accuracy'])
model1.fit(standard_train, Y_train, epochs=100)
```

```python
prediction = model1.predict([standard_test])
prect = []
for i in range(len(prediction)):
    prect.append(np.argmax(prediction[i]))
```

```python
accuracy(labels_test,prect)
confusion(labels_test,prect)
```

```python
#Second Requirement

tf.keras.backend.clear_session()
model2 = tf.keras.models.Sequential()
model2.add(tf.keras.layers.Flatten())
model2.add(tf.keras.layers.Dense(units=500, activation=tf.nn.relu))
model2.add(tf.keras.layers.Dense(units=10, activation=tf.nn.sigmoid))
model2.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mean_squared_error',metrics=['binary_accuracy'])
model2.fit(standard_train, Y_train, epochs=200)
```

```python
prediction = model2.predict([standard_test])
prect = []
for i in range(len(prediction)):
    prect.append(np.argmax(prediction[i]))
```

```python
accuracy(labels_test,prect)
confusion(labels_test,prect)
```
